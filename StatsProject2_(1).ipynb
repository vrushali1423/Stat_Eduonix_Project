1.Data Handling:
How would you handle missing values in a dataset? Describe at least two methods.

(1)Deletion Method:
Remove rows or columns with missing values if the percentage of missing data is low.

Example: Drop rows with NaN values using tools like dropna() in Python (Pandas).

(2)Imputation Method:
Replace missing values with the mean, median, or mode of the respective column.

Example: Use fillna() in Python to replace NaN values with the column mean.

2.Statistical Analysis:
What is a T-test, and in what scenarios would you use it? Provide an example based on sales data.

A T-test compares the means of two groups to determine if they are significantly different.

Example: Use a T-test to compare the average sales of two regions (Region A vs. Region B) to see if the difference is statistically significant.

Describe the Chi-square test for independence and explain when it should be used. How would you apply it to test the relationship between shipping mode and customer segment?

A Chi-square test evaluates whether two categorical variables are independent.

Example Application: Use it to determine if shipping mode (Standard, Express) is dependent on customer segment (Corporate, Consumer) by creating a contingency table and calculating the Chi-square statistic.

Purpose: Understand distribution, central tendency (mean, median), and variability (standard deviation).

Univariate Analysis: Analyzing a single variable, e.g., calculating average sales.

Bivariate Analysis: Examining relationships between two variables, e.g., correlation between sales and advertising expenses.
